#Import Libraries

```python
import pandas as pd
import seaborn as sns
```

#Import Data

```python
data = pd.read_csv('top_100_richest.csv', index_col = 0)
data.head()
```

#Explore Data

```python
len(data), len(data.dropna())
```

```python
data = data.dropna()
data.info()
```

#Data Cleaning

```python
data['net_worth'] = data['net_worth'].str.replace('$', '')
data['net_worth'] = data['net_worth'].str.replace('Billion', '')
data.head()
```

```python
data['net_worth'] = data['net_worth'].astype(float)
data.info()
```

```python
data = data.drop('bday', axis = 1)
data.head()
```

```python
data.describe()
```

```python
data.sort_values('net_worth', ascending = False).head(20)
```

```python
data.groupby('nationality').mean().sort_values('net_worth', ascending = False)
```

#Data Visualize

```python
sns.regplot(x = 'net_worth', y = 'age', data=data)
```

```python
sns.barplot(x = 'net_worth', y = 'nationality', data = data)
```

```python
sns.barplot(x = 'age', y = 'nationality', data = data)
```

```python
sns.displot(data['net_worth'], kde = True, bins = 20)
```

```python
sns.displot(data['age'], kde = True, bins = 20)
```
